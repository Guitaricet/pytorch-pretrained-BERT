```python
import re
import os

from tqdm import tqdm
from rusenttokenize import ru_sent_tokenize

LOADFROM = '/data/language_modeling/ru_lenta_y_wiki.txt'
SAVETO = '/data/preprocessed/bert/ru_lenta_y_wiki.txt'

os.makedirs(os.path.dirname(SAVETO), exist_ok=True)
```

```python
with open(LOADFROM) as f:
    for i, line in tqdm(enumerate(f), total=1463888+10000):
        if i > 2:
            break
        print(line)
        print('-'*100)
```

```python
def preprocess(text):
    text = re.sub(u'\xa0', ' ', text)
    text = '\n'.join(ru_sent_tokenize(text))
    text = text + '\n'
    return text
```

```python
# import warnings
import logging
logging.disable(logging.DEBUG)
logging.disable(logging.WARNING)
# warnings.simplefilter('ignore')
```

```python
with open(LOADFROM) as f:
    with open(SAVETO, 'w') as out:
        for i, line in tqdm(enumerate(f), total=2774951):
            out.write(preprocess(line))
```
